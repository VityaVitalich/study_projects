### - (2 pts) Докажите, что $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$.


Пусть rank(A) = r.

$\sqrt{<A,A>} \leq \sqrt(r) \|A\|_2$

$<A,A> \leq r\|A\|_2^2$

$tr(A^*A) \leq r \lambda_{max}(A^*A)$

Применяя SVD получим

$tr(A^*A) = tr(AA^*) = tr(U\Sigma V^* V \Sigma^* U^*) = tr(U\Sigma \Sigma^* U^*) = tr(\Sigma \Sigma^*) = \sum\limits_{i=1}^r \sigma_i \overline{\sigma_i}$, так как $\sigma_k = 0 \, \forall k>rank(A) = r$

В ту же очередь
$\lambda_{max}(A^*A) = \sigma_1 \overline{\sigma_1}$

Тогда 

$\sum\limits_{i=1}^r \sigma_i \overline{\sigma_i} \leq r (\sigma_1 \overline{\sigma_1})$, так как $\sigma_1 \geq \sigma_i \forall i$

$\blacksquare$

### - (2 pts) Получите сингулярное разложение и спектральную норму матрицы $A = uv^*$, где $u \in \mathbb{C}^m, v \in \mathbb{C}^n$

### - (3 pts) Докажите, что если последовательность унитарных матриц $\{U_k\}$ сходится к матрице $U$ по спектральной норме, то $U$ также унитарная матрица


Требуется доказать, что 
$U_kU_k^* = I \, \forall k \, \, \,  \lim\limits_k \|U_k\|_2 = \|U\|_2 \Rightarrow UU^* = I$


Для начала рассмотрим спектральную норму унитарной матрицы.

$||U||_2 = \displaystyle\sup_{x\neq0} \cfrac{||Ux||_2}{||x||_2}$

Заметим, что 

$||Ux||_2^2 = <Ux, Ux> = x^*U^*Ux = x^*x = ||x||_2^2$

Тогда спектральная норма унитарной матрицы равна 1

$\displaystyle\sup_{x\neq0} \cfrac{||Ux||_2}{||x||_2} = \displaystyle\sup_{x\neq0} \cfrac{||x||_2}{||x||_2} = 1$

Следовательно 

$\lim\limits_k \|U_k\|_2 = \lim\limits_k 1 = \|U\|_2$

Теперь покажем в обратную сторону
$\|U\|_2 = 1 \Rightarrow UU^* = I$


$\|U\|_2 = \displaystyle\sup_{x\neq0} \cfrac{||Ux||_2}{||x||_2} = 1 \Rightarrow ||Ux||_2 = ||x||_2 \Rightarrow ||Ux||_2^2 = ||x||_2^2 \Rightarrow x^*U^*Ux = x^*x \Rightarrow U^*U = I$

$\blacksquare$
